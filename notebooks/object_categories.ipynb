```python
import numpy as np
import matplotlib.pyplot as plt
from multimodal.object_categories_data_module import ObjectCategoriesDataModule
from multimodal.attention_maps import gradCAM, getAttMap, n_inv, imshow
```

```python
object_categories_dm = ObjectCategoriesDataModule()
object_categories_dm.prepare_data()
object_categories_dm.setup()

dataloader = object_categories_dm.test_dataloader(shuffle=True)
dataloader = iter(dataloader)
print(len(dataloader))
```

```python
batch = dataloader.next()
imgs, labels, labels_len, raw_label = batch
original_imgs = n_inv(imgs).squeeze(0)

print(raw_label)
plt.figure(figsize=(20, 15))
for i, original_img in enumerate(original_imgs):
    np_img = original_img.permute((1, 2, 0)).cpu().numpy()

    ax = plt.subplot(4, 4, i+1)
    plt.imshow(np_img)
    plt.xticks([])
    plt.yticks([])
```
